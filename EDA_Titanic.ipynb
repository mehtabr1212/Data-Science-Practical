```python
# Importing all necessary Libraries
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df=sns.load_dataset("titanic")
df.head()
```

```python
# Printing Shape Dimensions
df.shape
```

```python
# Descriptive summary
df.describe()
```

```python
# Checking null values
df.isnull().sum()
```

```python
df=df.drop('deck', axis=1)
df.head()
```

```python
df["age"] = df["age"].fillna(df["age"].mean())
df["embark_town"] = df["embark_town"].fillna(df["embark_town"].mode()[0])
df.head()
```

```python
df = df.rename(columns = {"sex": "Gender","fare": "Price"})
df.head()
```

```python
# Complete Columns Information about dataset
df.info()
```

```python
df.columns
```

```python
df.duplicated().sum()
```

```python
df.nunique()
```

# **Univariate Analysis**

Total number of child, man and woman paid the price for tickets:

```python
df.groupby('who').agg({'Price':'count'})
```

```python
df['pclass'].value_counts()
```

```python
sns.countplot(x='pclass',data=df, hue='who')
plt.title("Different Classes of ship with respect to gender")
plt.xlabel("Classes")
plt.ylabel("Count")
```

```python
df['who'].value_counts()
```

```python
who_counts=df['who'].value_counts()
plt.pie(who_counts, labels=who_counts.index, autopct = '%2.1f%%', colors=["orange", "skyblue", "red"])
plt.title("Distribution of Genders:")
plt.show()
```

# Resultant Observation from above

In Who column the "Maximum" dominated gender is Man

```python
sns.boxplot(y="age",data=df, hue="embark_town")
print("Max age: ", df.age.max(), "\nMin age: ", df.age.min())
```

# Resultant Observation from above

Highest number of gender who survived was female

```python
df.hist('age', bins= 50, color='orange')
plt.show()
```

Highest number of age is around 30

```python
sns.boxplot(y='Price', data=df, hue='class')
print("Max fare: ", df.Price.max(), "\nMin fare: ", df.Price.min())
```

# **Bivariate Analysis**

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.stripplot(x='survived', y='age', data=df, palette='rocket_r', hue='Gender')

total_survived = df['survived'].value_counts()[1]
total_not_survived = df['survived'].value_counts()[0]

print("Total People Survived:", total_survived)
print("Total People Not Survived:", total_not_survived)

plt.title("Survival Count by Gender")
plt.xlabel("Gender")
plt.ylabel("Age")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# line plot
sns.lineplot(x='age', y='Price', data=df, estimator='mean')
plt.title('Average Fare Paid by Age')
plt.xlabel('Age')
plt.ylabel('Average Price')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x='age', y='Price', data=df)
plt.title('Age vs Fare')
plt.show()

sns.boxplot(x='pclass', y='Price', data=df)
plt.title('Pclass vs Fare')
plt.show()
```
